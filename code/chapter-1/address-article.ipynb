## 주소 보도자료 크롤링
- 송채은
- 2024-03-11

- 주소정보 누리집
    - 고객지원 > 언론보도
    - 도로명주소 소개 > 홍보자료

```python
from bs4 import BeautifulSoup
import requests
import re
import os
import sys
import time
import random
import json
import pandas as pd
import numpy as np
#tqdm
from tqdm import tqdm
today = time.strftime("%Y-%m-%d", time.localtime())
print(today)
```

```python
# 주소정보 누리집 '고객지원 > 언론보도'
page_num = 1
meta_list = []
for page_num in range(1, 12):
    url = f'https://www.juso.go.kr/notice/NewsBoardList.do?currentPage={page_num}&countPerPage=10&searchType=&keyword='
    html = requests.get(url).text
    soup = BeautifulSoup(html, 'html.parser')
    doc_list = soup.select('table.tbl-black.board > tbody > tr')
    for doc in doc_list:
        no = doc.select_one('td.no').get_text().strip()
        title = doc.select_one('td.title > a').get_text().strip()
        date = doc.select_one('td.date').get_text().strip()
        viewCnt = doc.select_one('td.viewCnt').get_text().strip()
        doc_url = doc.select_one('td.title > a')['href']
        print(no, title, date, viewCnt)
        meta_list.append([no, title, date, viewCnt, doc_url])
```

```python
df = pd.DataFrame(meta_list, columns=['no', 'title', 'date', 'viewCnt', 'doc_url'])
df
```

```python
df['doc_url'] = 'https://www.juso.go.kr' + df['doc_url']
```

```python
df
```

```python
df['content'] = None
df['ref_url'] = None
for i in range(len(df)):
    url = df['doc_url'][i]
    html = requests.get(url).text
    soup = BeautifulSoup(html, 'html.parser')
    content = soup.select_one('div.table.table_view > table > tbody > tr:nth-child(4) > td').get_text().strip()
    try:
        # '보로자료 바로가기' 인 a 태그의 href 속성값을 가져온다
        ref_url = soup.select_one('div.table.table_view > table > tbody > tr:nth-child(4) > td a')['href']
    except:
        ref_url = None
    df['content'][i] = content
    df['ref_url'][i] = ref_url
```

```python
df
```

```python
df.to_csv(f'./주소정보 누리집_언론보도_{today}.csv', index=False, encoding='utf-8-sig')
```

## 네이버 블로그 -  행정안전부 
- 주소 관련: https://blog.naver.com/PostSearchList.naver?blogId=mopaspr&categoryNo=0&SearchText=%EC%A3%BC%EC%86%8C&orderBy=sim&term=&startDate=&endDate=&range=all&cpage=1

```python
page_num = 1
meta_list = []

for page_num in tqdm(range(1, 39), total=39):
    url = f'https://blog.naver.com/PostSearchList.naver?blogId=mopaspr&categoryNo=0&SearchText=%EC%A3%BC%EC%86%8C&orderBy=sim&term=&startDate=&endDate=&range=all&cpage={page_num}'
    html = requests.get(url).text
    soup = BeautifulSoup(html, 'html.parser')
    title_list = soup.select('a.s_link')
    for title in title_list:
        doc_url = title['href']
        title = title.get_text().strip()
        meta_list.append([title, doc_url])
```

```python
df = pd.DataFrame(meta_list, columns=['title', 'doc_url'])
df
```

```python
df['content'] = None

for idx, row in df.iterrows():
    # 'https://blog.naver.com/mopaspr?Redirect=Log&logNo=223365752088&from=postView'
    log_no = re.search('logNo=(\d+)', row['doc_url']).group(1)
    url = f'https://blog.naver.com/PostView.naver?blogId=mopaspr&logNo={log_no}&redirect=Dlog&widgetTypeCall=true&directAccess=false'
    html = requests.get(url).text
    soup = BeautifulSoup(html, 'html.parser')
    try:
        content = soup.select_one('div.se-main-container').get_text().strip()
    except:
        try:
            content = soup.select_one('div#postViewArea').get_text().strip()
        except:
            # 반복문 중단
            print('Error', idx, row['doc_url'])
            #에러문 출력
            print(sys.exc_info())
            break
    df['content'][idx] = content
```

```python
df
```

```python
df = df[['title', 'content', 'doc_url']]
df.to_csv(f'./네이버블로그_행정안전부_주소_{today}.csv', index=False, encoding='utf-8-sig')
```
